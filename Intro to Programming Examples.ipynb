```python
import matplotlib.pyplot as plt
```

```python
def create_circle(canvas, x_pos, y_pos, radius_parameter, color_parameter='black'):
    circle= plt.Circle((x_pos, y_pos), radius=radius_parameter, color=color_parameter)
    
    canvas.add_patch(circle)
    plt.axis('scaled')
    
    
```

```python
canvas=plt.gca()
create_circle(canvas, 0, 0, 5)
create_circle(canvas, 3, 3, 4, color_parameter='red')
```

```python
canvas =plt.gca()
for i in range(10):
    for j in range(10):
        if i == j:
            create_circle(canvas, i, j, 1, color_parameter='red')
```
