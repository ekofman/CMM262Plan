```python
import matplotlib.pyplot as plt
```

# Functions

```python
def create_circle(canvas, x_pos, y_pos, radius_parameter, color_parameter='black'):
    circle= plt.Circle((x_pos, y_pos), radius=radius_parameter, color=color_parameter)
    
    canvas.add_patch(circle)
    plt.axis('scaled')
```

```python
canvas=plt.gca()
create_circle(canvas, 0, 0, 5)
create_circle(canvas, 3, 3, 4, color_parameter='red')
```

### Make your own function!

```python
canvas=plt.gca()

def create_snowman(canvas, color_parameter='#bbbbbb'):
    create_circle(canvas, 0, 0, 5, color_parameter=color_parameter)
    create_circle(canvas, 0, 5, 4, color_parameter=color_parameter)
    create_circle(canvas, 0, 10, 3, color_parameter=color_parameter)
    create_circle(canvas, 1, 10, .5, color_parameter='black')
    create_circle(canvas, -1, 10, .5, color_parameter='black')
    create_circle(canvas, 0, 9, .5, color_parameter='orange')


create_snowman(canvas)
```

# If/else statements

```python
import random
x = random.random()

if x > .5:
    color = 'yellow'
else:
    color = 'blue'

canvas=plt.gca()
create_circle(canvas, 0, 0, 5, color_parameter=color)
```

# For loops

```python
canvas =plt.gca()
for i in range(10):
    for j in range(10):
        if i == j:
            create_circle(canvas, i, j, 1, color_parameter='red')
```
